```python
import re
import chardet
```

```python
ftow = open('1_cleaned', 'w')
with open('1.txt', mode = 'r', encoding = 'GBK') as f:
    for line in f:
        string = re.sub("[\s+\.\!\/_,$%^*(+\"\']+|[”“+——、~@#￥%……&*（）]+", "",line) # 去特殊字符
        string = re.sub("[。！？：，]+", "\n", string) # 换行
        if not re.search("[a-zA-Z]+|===", string) == None:
            continue
        ftow.write(string)   
        print(string)
ftow.close()
```
